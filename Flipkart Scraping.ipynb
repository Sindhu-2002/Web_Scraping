```python
Product_name = []
Prices = []
Description= []
```

```python
#fetching data specifically from a page
url = "https://www.flipkart.com/search?q=mobiles+under+50000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=1"
r = requests.get(url)
soup = BeautifulSoup(r.text,"lxml")

names = soup.find_all("div",class_ = "_4rR01T")
print(names)
```

```python
#extracting names
names = soup.find_all("div",class_ = "_4rR01T")
for i in names:
    name = i.text
    Product_name.append(name)

print(Product_name) 
```

```python
#similarly extract prices
prices = soup.find_all("div",class_ = "_30jeq3 _1_WHN1")
for i in prices:
    name = i.text
    Prices.append(name)
print(Prices)
```

```python
#Extracting Description
desc = soup.find_all("ul",class_ = "_1xgFaf")
for i in desc:
    name = i.text
    Description.append(name)
print(Description)
```

```python
#check for lengths of all attributes to find unnecessary details

len(Product_name) #24
len(Prices) #24
len(Description) #24
```

# Creating a Dataframe

```python
for i in range(1,40):
    url = "https://www.flipkart.com/search?q=mobiles+under+50000&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page="+str(i)
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "lxml")
    #box = soup.find("div", class_ = "_1YokD2 _3Mn1Gg")
    
    names = soup.find_all("div", class_ = "_4rR01T")
    for i in names:
        name = i.text
        Product_name.append(name)
        
    prices = soup.find_all("div", class_ = "_30jeq3 _1_WHN1")
    for i in prices:
        name = i.text
        Prices.append(name)
        
    desc = soup.find_all("ul",class_ = "_1xgFaf")
    for i in desc:
        name = i.text
        Description.append(name)
    
        
```

```python
import pandas as pd
df = pd.DataFrame({"Product Name":Product_name,"Price":Prices,"Description":Description})
print(df)
```
